```python
# Removing NaN / NULL vales from the data
df.dropna(axis = 0, inplace = True)
print(df.shape)
```

```python
# Target Variable
df['TenYearCHD'].value_counts()
```

```python
import plotly.express as px
fig = px.pie(df, names='TenYearCHD', title='Distribution of TenYearCHD')
fig.show()
```

```python
fig = px.scatter(df, x='age', y='sysBP', color='TenYearCHD', title='Scatter Plot of Age vs. Systolic Blood Pressure')
fig.show()
```

```python
X = df.drop('TenYearCHD', axis=1)
y = df['TenYearCHD']
```

```python
# Assigning the data for training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=21)
```

```python
#Applying the ML model - Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
#Training the data
logreg.fit(X_train, y_train)
```

```python
#Testing the data
y_pred = logreg.predict(X_test)
```

```python
# Getting the Confusion Matrix and Classification Report
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(y_pred)
print("Confusion Matrix is:\n",cm)
print("Classification Report is:\n\n",classification_report(y_test,y_pred))
```
