```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

cmap = ListedColormap(['red', 'green', 'blue'])

plt.scatter(x, y, c=colors, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
# Customizing Colormaps
#You can modify existing colormaps by changing properties like color limits.

import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl

data = np.random.rand(4, 4)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 4))

ax1.imshow(data)
ax1.set_title("Default colormap")

mpl.rc('image', cmap='plasma')

ax2.imshow(data)
ax2.set_title("Modified colormap")

plt.show()
```
